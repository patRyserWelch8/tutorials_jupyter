### Import GDP 

Source: Kaggle 
[URL](https://www.kaggle.com/datasets/rinichristy/countries-gdp-19602020)

```python
file_name = "data/Countries_GDP_1960_2020.csv"
GDP_df_wide = pd.read_csv(file_name)
```

```python
GDP_df_wide.columns
```

```python
GDP_df_wide.head()
```

Convert the year as column into rows



```python
years = np.arange(1960,2020)
years = list(map(str,years))

columns_name = ["country_name","country_code","gdp", "year"]

GDP_df_long = pd.DataFrame()
print(GDP_df_long)

for year in years:
    columns = ['Country Name', 'Country Code', year]
    data    = GDP_df_wide[columns].copy(deep=True)
    year_values = pd.Series(np.repeat(int(year), len(data)))
    data['year'] = year_values
    data.columns = columns_name
    GDP_df_long = GDP_df_long.append(data, ignore_index=True)

GDP_df_long
```

### Import Air polluation

Source: Kaggle [URL](https://www.kaggle.com/datasets/kweinmeister/pm25-global-air-pollution-20102017)

```python
file_name = "data/PM2_5_Global_Air_Pollution_2010_2017.csv"
air_pol_df_wide = pd.read_csv(file_name)
air_pol_df_wide.head()
```

Convert years into rows

```python
years = np.arange(2010,2017)
years = list(map(str,years))

columns_name = ["country_name","country_code","pm2_5", "year"]

air_pol_df_long = pd.DataFrame()


for year in years:
    columns = ['Country Name', 'Country Code', year]
    data    = air_pol_df_wide[columns].copy(deep=True)
    year_values = pd.Series(np.repeat(int(year), len(data)))
    data['year'] = year_values
    data.columns = columns_name
    air_pol_df_long = air_pol_df_long.append(data, ignore_index=True)

air_pol_df_long
```

## Merge datasets into one

### Find discrepancies between air and GDP

```python
air_pol_df_long
```

```python
air_country_codes = set(pd.unique(air_pol_df_long["country_code"]))
air_country_codes
```

```python
GDP_df_long
```

```python
gdp_country_codes = set(pd.unique(GDP_df_long["country_code"]))
gdp_country_codes 
```

```python
intersection = air_country_codes.intersection(gdp_country_codes)
intersection = list(intersection)
intersection 
```

```python
air_pol_df = air_pol_df_long.copy(deep = True)
air_pol_df = air_pol_df.loc[air_pol_df['country_code'].isin(intersection)]
air_pol_df
```

```python
gdp_df = GDP_df_long.copy(deep = True)
gdp_df = gdp_df.loc[gdp_df['country_code'].isin(intersection)]
gdp_df

```

#### merge both datasets 

```python
gdp_air_df = gdp_df.merge(air_pol_df, how="inner", on=['country_name','country_code', 'year'])
gdp_air_df
```

```python
gdp_air_df.to_csv(file_path)
```
