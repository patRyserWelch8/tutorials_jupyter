The factorial function (symbol: !) says to multiply all whole numbers from our chosen number down to 1.

## How can we implement factorial in Python

### Option 1
Let's use a function and while loop


```python
def factorial(number: int) -> int:
    """
    compute the factorial of a number
    input a number as integer
    precondition: number is greater than 1.
    postcondition: number!
    """
    # init factors
    factor:int  = number
    outcome:int = 1
    
    # iteratively multiply factorial
    while factor >= 1:
        outcome = outcome * factor
        factor = factor - 1
        
    return outcome
    
```

### option 2
Let's use a function and for loop

```python
def factorial(number: int) -> int:
    """
    compute the factorial of a number
    input a number as integer
    precondition: number is greater than 1.
    postcondition: number!
    """
   
    outcome:int = 1
    
    # iteratively multiply factorial
    for factor in range(number, 1, -1):
        outcome = outcome * factor
        factor = factor - 1
        
    return outcome
```

## option 4

[The Sieve of Eratosthenes](https://cp-algorithms.com/algebra/prime-sieve-linear.html)

The idea behind is this: 
A number is prime, if none of the smaller prime numbers divides it.

Since we iterate over the prime numbers in order, we already marked all numbers, who are divisible by at least one of the prime numbers, as divisible. Hence if we reach a cell and it is not marked, then it isn't divisible by any smaller prime number and therefore has to be prime.



```python
def search_primes(number:int) -> []:
    """
    returns the prime factorisation of a number
    precondition: number > 1
     postcondition: returns at least of vector of primes >= 1
    """
     
    list_numbers = [None] * (number+1)
    primes = []
    for i in range(2, number+1):
        if not list_numbers[i]:
            primes.append(i)
            list_numbers[i] = i
        for prime in primes:
            if prime > list_numbers[i] or i * prime > number:
                break
            list_numbers[i * prime] = prime
    return primes
```

```python
def factorial(number:int) -> int:
    """Return the prime factorization of n! as list of 
    (prime, exponent) pairs."""
    
    outcome = 1
    primes  = search_primes(number)
    
    # Find prime exponents for n! and 
    for prime in primes:
        exponent = 0
        quotient = number // prime
        while quotient > 0:
            exponent += quotient
            quotient = quotient // prime
        outcome = outcome * (prime ** exponent)
    return outcome
```
